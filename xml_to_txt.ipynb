```python
# 한 xml 파일에 대한 전체 코드 for 문으로 생성

import os
import json

classes = ['wheelchair','truck','tree_trunk','traffic_sign','traffic_light','traffic_light_controller','table',
          'stroller','stop','scooter','potted_plant','power_controller','pole','person','parking_meter',
          'movable_signage','motorcycle','kiosk','fire_hydrant','dog','chair','cat',
          'carrier','car','bus','bollard','bicycle','bench','barricade']

classes_dict = {class_name : i for i,class_name in enumerate(classes)} 


for k in range(0,len(json_data['annotations']['image'])):
    json_dir = './json/data'+str(k)+'.json'  # JSON file path
    out_dir = './txt/'  #  Output TXT file path
    
    def change_classes(file_name,classes_dict):
        str_to_num = []
        with open(file_name) as f:
            lines = f.readlines()
            for line in lines:
                for key,value in enumerate(classes_dict):
                    line = line.replace(value,str(key))
                str_to_num.append(line)
            f.close()
            
        with open(file_name,'w') as f:
            f.writelines(str_to_num)
            f.close()

    def main():
        with open(json_dir, 'r', encoding='utf-8') as load_f:
            content = json.load(load_f)
        #  Cyclic processing
        tmp = content['@name'].split('.')
        filename = out_dir + tmp[0] + '.txt'


        
        for i in range(0,len(content['box'])+1):
            if os.path.exists(filename):
                x = (float(content['box'][i-1]['@xtl']) + float(content['box'][i-1]['@xbr'])) / 2 / float(content['@width'])
                y = (float(content['box'][i-1]['@ytl']) + float(content['box'][i-1]['@ybr'])) / 2 / float(content['@height'])
                w = (float(content['box'][i-1]['@xbr']) - float(content['box'][i-1]['@xtl'])) / float(content['@width'])
                h = (float(content['box'][i-1]['@ybr']) - float(content['box'][i-1]['@ytl'])) / float(content['@height'])
                fp = open(filename, mode="r+", encoding="utf-8")
                file_str = content['box'][i-1]['@label'] + ' ' + str(round(x, 6)) + ' ' + str(round(y, 6)) + ' ' + str(round(w, 6)) + \
                           ' ' + str(round(h, 6))
                line_data = fp.readlines()
        


                if len(line_data) != 0:
                    fp.write('\n' + file_str)
                else:
                    fp.write(file_str)
                fp.close()
            
            #  Create a file without existence
            else:
                fp = open(filename, mode="w", encoding="utf-8")
                fp.close()
        
        change_classes(filename,classes_dict)

    if __name__ == '__main__':
        main()
 
```
